### **> Anaconda Setup & Python Environment**

### **Objective**<br>
To install Anaconda, create a Python virtual environment, and install basic data-science packages.

### **Tools Used**<br>
Windows OS
Anaconda Distribution
Visual Studio Code
Python 3.11

### **Procedure**<br>
**Anaconda Installation:**<br>
Anaconda was downloaded from the official website and installed using Just Me option without adding it to system PATH.
Website: https://www.anaconda.com/

**VS Code Installation:**<br>
Visual Studio Code was installed from the official website for Python development.
Website: https://code.visualstudio.com/

**Create Virtual Environment:**<br>
conda create --name mycourse python=3.11 -y
Activate Environment
conda activate mycourse

**Install Required Packages:**<br>
conda install numpy pandas matplotlib jupyter -y
conda install -c conda-forge scikit-learn seaborn -y
conda install ipykernel -y

### **conclusion**
The Python environment is now ready for data analysis and programming tasks.
