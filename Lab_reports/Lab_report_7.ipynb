### **>Monitoring a Solar Home System (Dictionary in Python)**
You are working in a university electronics lab. You are asked to monitor a small solar home system installed on the lab roof. The system has:<br> a) A solar panel<br> b) A battery<br> c) A load controller<br> d) A temperature sensor<br> Each minute, the system sends a small dictionary of readings<br> I. Create a dictionary named reading1 containing the 5 fields shown in the scenario.<br> II. From reading1, print:<br> i. The solar power<br> ii. The battery voltage<br> iii. The load status<br> III. Print how many fields the dictionary contains after your modifications.<br> IV. Remove the "panel_current" field.<br> V. Get Keys, Values, and Items Print:<br> i. All keys ii. All values<br> iii. All key-value pairs<br> iv. Use .keys(), .values(), and .items().

### **Objective**
The objective of this lab is to use Python dictionaries to store, access, modify, and analyze sensor data from a small solar home system installed in a university electronics lab.

### **Topics Covered**
Creating dictionaries<br>
Accessing dictionary values<br>
Modifying dictionary data<br>
Removing dictionary fields<br>
Using .keys(), .values(), and .items()<br>

```python
reading1 = {
    "solar_power": 250,
    "panel_current": 8.5,
    "battery_voltage": 12.6,
    "load_status": "ON",
    "temperature": 35
}

print(reading1["solar_power"])
print(reading1["battery_voltage"])
print(reading1["load_status"])

print(len(reading1))

reading1.pop("panel_current")

print(reading1.keys())
print(reading1.values())
print(reading1.items())
```
