### **> Use of sum() Function in Python**
sum: sum() adds up numbers in a list.<br>
i.	The power consumed by individual components is listed in a list. Find the total power consumed by adding readings.<br>
ii.	A list contains the values of a series of connected resistances. Find its equivalent resistance using the sum.<br>
Example:<br>
nums = [1, 2, 3]<br>
print(sum(nums))<br>
Output:<br>
6<br>


```python
power = [10, 20, 30, 40]
print(sum(power))

resistances = [5, 10, 15]
print(sum(resistances))
```

```python
# Example:
nums = [1, 2, 3]
print(sum(nums))
# Output:
# 6
```
