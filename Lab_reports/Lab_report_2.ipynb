### **> Find Total Impedance**
i. Add the three impedances. 
<br>
ii. Print real part (resistive) and imaginary part (reactive). <br>
iii. Print magnitude and phase angle (use abs() and cmath.phase()). 

```python
import cmath

z1 = 3 + 4j
z2 = 5 - 2j
z3 = 1 + 6j

z_total = z1 + z2 + z3

print(z_total)
print(z_total.real)
print(z_total.imag)
print(abs(z_total))
print(cmath.phase(z_total))
```
