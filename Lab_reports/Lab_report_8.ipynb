### **> Using Lambda Functions for Electrical Lab Calculations**
You’re writing small analysis tools for an electrical lab. You need quick one-line functions to calculate adjusted power readings, filter noisy sensor data, sort components by rating, and combine multiple electrical values. Each task builds a specific mini-tool using Python lambda functions.<br>
Task 1 — Basic lambda for a simple electrical formula<br>
Write a lambda that takes a voltage value V and returns V² + 2.
Test it with V = 5.<br>
(Represents a toy “adjusted voltage metric.”)<br>
Task 2 — Lambda with two electrical parameters<br>
Write a lambda that takes two resistances R1 and R2 and returns their equivalent parallel resistance using:

$$
R_{\rm eq} = \frac{R_1 \, R_2}{R_1 + R_2}
$$

Test it with:<br>
	R1 = 100 Ω<br>
	R2 = 200 Ω


### **Objective**
The objective of this lab is to understand and apply Python lambda functions to perform quick mathematical operations related to electrical laboratory analysis.<br>

### **Topics Covered**
Lambda functions<br>

```python
# task 1
adjusted_voltage = lambda V: V**2 + 2
print(adjusted_voltage(5))
# task 2
parallel_resistance = lambda R1, R2: (R1 * R2) / (R1 + R2)
print(parallel_resistance(100, 200))

```

```python
# task 1
adjusted_voltage = lambda V: V**2 + 2
print(adjusted_voltage(5))
```
