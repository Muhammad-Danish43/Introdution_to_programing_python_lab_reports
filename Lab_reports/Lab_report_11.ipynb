### **> Data Transformation Using Dictionary Comprehensions in Python**
You are helping a small learning center clean and organize small classroom datasets.<br> They give you short dictionaries and want quick transformations using comprehensions only, because they want the code short and readable.<br> Create dictionary from a list<br> Given:<br> names = ["Ali", "Sara", "Bilal"]<br> Create a dictionary where each name maps to its length.<br> {key: value for item in iterable}<br> Filter a dictionary using comprehension<br> Given:<br> scores = {"Ali": 55, "Sara": 82, "Bilal": 40, "John": 90}<br> Create a new dictionary containing only the students who scored 60 or above.<br> {k: v for k, v in dict.items() if condition}<br> Swap keys and values<br> Given:<br> pairs = {"a": 1, "b": 2, "c": 3}<br> Use a dictionary comprehension to swap keys and values.<br> Transform dictionary values<br> Given:<br> prices = {"pen": 10, "book": 50, "bag": 100}

**Objective**
To perform quick and readable transformations on small datasets using Python dictionary comprehensions and display the results.

**Topics Covered**
Dictionary Comprehension<br>
Creating Dictionaries<br>
Filtering Dictionaries<br>
Swapping Keys and Values<br>
Transforming Dictionary Values<br>

```python
names = ["Ali", "Sara", "Bilal"]
name_lengths = {name: len(name) for name in names}
print(name_lengths)
```

```python
scores = {"Ali": 55, "Sara": 82, "Bilal": 40, "John": 90}
passed_students = {k: v for k, v in scores.items() if v >= 60}
print(passed_students)
```

```python
pairs = {"a": 1, "b": 2, "c": 3}
swapped = {v: k for k, v in pairs.items()}
print(swapped)
```

```python
prices = {"pen": 10, "book": 50, "bag": 100}
discounted_prices = {k: v * 0.9 for k, v in prices.items()}
print(discounted_prices)
```
