### **> Calculation of Square of Current Readings Using Python**
Current readings are stored in a list. Engineers want to calculate the square of each value for power loss calculation.<br>
Print the square of each current reading using formula I².<br>
	Make a list of current values, loop and multiply each by itself.<br>


### **Objective**
To calculate and print the square of each current reading stored in a list using the formula I².

### **Topics Covered**
Python Lists<br>
For Loop<br>
Arithmetic Operations<br>
Output Formatting<br>

```python
current_values = [2, 4, 6, 8, 10]

for I in current_values:
    print(I * I)
```
