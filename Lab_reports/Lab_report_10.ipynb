### **> Data Transformation Using List Comprehensions in Python**
You are helping a small learning center clean and organize small classroom datasets.<br>
They give you short lists and dictionaries and want quick transformations using comprehensions only, because they want the code short and readable.<br>
**A. List Comprehension Tasks**<br> 
**Task 1:**  Basic list comprehension<br>
Create a new list of numbers from 0 to 9 using a list comprehension, not a loop.<br>
basic syntax [expression for item in iterable]<br>
**Task 2:** Filtering items with if<br>
Given:<br>
nums = [1, 2, 3, 4, 5, 6]<br>
Use a list comprehension with an if clause to keep only the even numbers.<br>
 [expr for item in iterable if condition]<br>
**Task 3**:  Conditional expression inside comprehension<br>
Given:<br>
nums = [1,2,3,4]<br>
Create a new list where even numbers become "even" and odd numbers become "odd".<br>
expr_if_true if condition else expr_if_false inside comprehension<br>
**Task 4:** Simple nested list comprehension<br>
Given a list of lists:<br>
data = [[1,2], [3,4]]<br>
Flatten it into a single list using a nested list comprehension.<br>


### **Objective**
To perform basic data transformations on small datasets using Python list comprehensions for concise and readable code.

### **Topics Covered**
List Comprehension<br>
Conditional Filtering<br>
Conditional Expressions<br>
Nested List Comprehension<br>

```python
# task 1
numbers = [x for x in range(10)]
print(numbers)
```

```python
# task 2
nums = [1, 2, 3, 4, 5, 6]
evens = [x for x in nums if x % 2 == 0]
print(evens)
```

```python
# task 3
nums = [1, 2, 3, 4]
result = ["even" if x % 2 == 0 else "odd" for x in nums]
print(result)
```

```python
# task 4
data = [[1, 2], [3, 4]]
flat_list = [x for sublist in data for x in sublist]
print(flat_list)
```

```python
# all in one 
numbers = [x for x in range(10)]
nums = [1, 2, 3, 4, 5, 6]
evens = [x for x in nums if x % 2 == 0]
nums = [1, 2, 3, 4]
result = ["even" if x % 2 == 0 else "odd" for x in nums]
data = [[1, 2], [3, 4]]
flat_list = [x for sublist in data for x in sublist]
print(numbers)
print(evens)
print(result)
print(flat_list)
```
