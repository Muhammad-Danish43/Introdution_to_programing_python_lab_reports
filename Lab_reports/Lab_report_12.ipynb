### **Task 1: Handle Invalid User Input**
**Problem Statement**<br>
The program asks the user to enter total marks and obtained marks. The user may enter invalid input such as letters or symbols.

**Objective**<br>
To prevent the program from crashing and display an error message when non-numeric input is entered.

**Topics Covered**<br>
try<br>
except<br>
ValueError<br>

```python
try:
    total_marks = float(input("Enter total marks: "))
    obtained_marks = float(input("Enter obtained marks: "))
    print("Input accepted successfully")
except ValueError:
    print("Error: Please enter numeric values only")
```

### **Task 2: Handle Division by Zero**
**Problem Statement**<br>
If total marks are entered as zero, calculating percentage will cause a runtime error.

**Objective**<br>
To safely handle division by zero using exception handling.

**Topics Covered**<br>
ZeroDivisionError<br>
Multiple except blocks

```python
try:
    total_marks = float(input("Enter total marks: "))
    obtained_marks = float(input("Enter obtained marks: "))
    percentage = obtained_marks / total_marks
    print("Percentage:", percentage)
except ValueError:
    print("Error: Invalid numeric input")
except ZeroDivisionError:
    print("Error: Total marks cannot be zero")
```

**Task 3: Validate Logical Conditions Using raise**
**Problem Statement**<br>
Obtained marks must not be negative and must not be greater than total marks.

**Objective**<br>
To manually raise an exception when logical conditions are violated.

**Topics Covered**<br>
raise<br>
Exception

```python
try:
    total_marks = float(input("Enter total marks: "))
    obtained_marks = float(input("Enter obtained marks: "))

    if obtained_marks < 0 or obtained_marks > total_marks:
        raise Exception("Obtained marks must be between 0 and total marks")

    print("Marks are valid")
except Exception as e:
    print("Error:", e)
```

### **Task 4: Execute Code Only When No Error Occurs**
**Problem Statement**<br>
Display percentage and pass/fail status only when no error occurs.

**Objective**<br>
To execute result calculation only if input and validation are successful.

**Topics Covered**<br>
try<br>
except<br>
else

```python
try:
    total_marks = float(input("Enter total marks: "))
    obtained_marks = float(input("Enter obtained marks: "))

    if obtained_marks < 0 or obtained_marks > total_marks:
        raise Exception("Invalid marks entered")

    percentage = (obtained_marks / total_marks) * 100
except Exception as e:
    print("Error:", e)
else:
    print("Percentage:", percentage)
    if percentage >= 40:
        print("Status: Pass")
    else:
        print("Status: Fail")
```

### **Task 5: Guarantee Program Completion Using finally**
**Problem Statement**<br>
The program must display a completion message regardless of any error.

**Objective**<br>
To ensure program completion message is always displayed.

**Topics Covered**<br>
finally

```python
try:
    total_marks = float(input("Enter total marks: "))
    obtained_marks = float(input("Enter obtained marks: "))

    if total_marks == 0:
        raise ZeroDivisionError

    if obtained_marks < 0 or obtained_marks > total_marks:
        raise Exception("Invalid marks")

    percentage = (obtained_marks / total_marks) * 100
except ValueError:
    print("Error: Numeric input required")
except ZeroDivisionError:
    print("Error: Total marks cannot be zero")
except Exception as e:
    print("Error:", e)
else:
    print("Percentage:", percentage)
finally:
    print("Result processing completed.")
```
