### **Task 1: Create and Write Energy Data to a CSV File**
**Problem Statement**<br>
A CSV file is required to store daily electricity consumption data in an organized format.

**Objective**<BR>
To create a CSV file and write the header and daily energy consumption records.

**Topics Covered**<br>
File handling<br>
CSV writing<br>
Writing data to a file

```python
import csv

file = open("energy_data.csv", "w", newline="")
writer = csv.writer(file)

data = [
    ["Date", "Energy Consumed (kWh)"],
    ["2025-01-01", 12.5],
    ["2025-01-02", 14.2],
    ["2025-01-03", 10.8]
]

writer.writerows(data)

file.close()
print("Energy data written successfully")
```

**Task 2: Read and Display CSV File Data**
**Problem Statement**<br>
The stored energy consumption data must be read from the CSV file and displayed clearly.

**Objective**<br>
To read data from a CSV file and display all records.

**Topics Covered**<br>
File reading<br>
CSV reader<br>
Displaying file data

```python
import csv

file = open("energy_data.csv", "r")
reader = csv.reader(file)

for row in reader:
    print(row)

file.close()
```

### **Task 3: Append New Energy Records**
**Problem Statement**<br>
New daily energy consumption records need to be added without deleting existing data.

**Objective**<br>
To append new records to an existing CSV file while keeping the correct format.

**Topics Covered**<br>
File append mode<br>
CSV writing<br>
Data persistence

```python
file = open("energy_data.csv", "a")

file.write("2025-01-04,13.6\n")
file.write("2025-01-05,11.9\n")

file.close()
print("New energy records added successfully")
```
